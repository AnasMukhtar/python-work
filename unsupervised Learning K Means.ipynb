## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

## importing the dataset

```python
flower = sns.load_dataset('iris')
```

```python
flower.tail(20)
```

## Remove the Target Feature, we will deal

```python
descriptive_feature = flower.iloc[:,:-1].values
descriptive_feature
```

## finding the optimum number of clusters

```python
from sklearn.cluster import KMeans
WCSS = []
for i in range(1 , 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++' , max_iter = 300, n_init = 10, random_state = 42)
    kmeans.fit(descriptive_feature)
    WCSS.append(kmeans.inertia_)
plt.plot(range(1 ,11), WCSS)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## Apply K-Mean Clustring

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++' , max_iter = 300, n_init = 10, random_state = 42)
y_prediction = kmeans.fit_predict(descriptive_feature)
```

```python
y_prediction
```

## visualizing the clusters

## Ploting the basis of sepel feature

```python
plt.scatter(descriptive_feature[y_prediction == 0,0] ,descriptive_feature[y_prediction == 0,1], s = 100, c = 'red', label = 'cluster 1')
plt.scatter(descriptive_feature[y_prediction == 1,0] ,descriptive_feature[y_prediction == 1,1], s = 100, c = 'orange', label = 'cluster 2')
plt.scatter(descriptive_feature[y_prediction == 2,0] ,descriptive_feature[y_prediction == 2,1], s = 100, c = 'blue', label = 'cluster 3')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker="X", s = 200, c = 'black', label = 'centroids')

plt.legend()
plt.show()
```

## DB SCAN with data point in Circular Pattern

```python
import math
import matplotlib.pyplot as plt
import matplotlib
```

## Create Data points in a form of Circle

```python
np.random.seed(42)
def PointsInCircum(r,n=100):
    return[(math.cos(2*math.pi/n*x)*r+np.random.normal(-30,30),math.sin(2*math.pi/n*x)*r+np.random.normal(-30,30))for x in range(1,n+1)]
```

```python
PointsInCircum(9,700)
```

```python
# creating data points in the form of a circule
df=pd.DataFrame(PointsInCircum(500,1000))
df=df.append(PointsInCircum(300,700))
df=df.append(PointsInCircum(100,300))
df=df.append([(np.random.randint(-600,600),np.random.randint(-600,600)) for i in range(300)])
```

## Plotting data points

```python
plt.figure(figsize=(10, 10))
plt.scatter(df[0],df[1],s=15,color='black')
plt.title('Dataset',fontsize=20)
plt.ylabel = 'Actual output'
plt.xlabel = 'predicted output'
plt.show()
```

## Clusterring on the basis of K-Means

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters = 4,random_state=42)
k_means.fit(df[[0,1]])
df['KMeans_labels']=k_means.labels_


```

```python
# ploting resulting clusters
colors=['purple','red','yellow','black']
plt.figure(figsize=(10, 10))
plt.scatter(df[0], df[1], c=df['KMeans_labels'],cmap=matplotlib.colors.ListedColormap(colors),s=15)
plt.title('KMeans Clustering',fontsize=20)
plt.ylabel = 'Actual Output'
plt.xlabel = 'predicted output'
plt.show()
```

## Clusterring on the basis of Aglomerative clustering

```python

from sklearn.cluster import AgglomerativeClustering
clusters= AgglomerativeClustering(n_clusters=3,affinity='euclidean')
clusters.fit(df[[0,1]])
df['HR_labels']=clusters.labels_
```

```python
# ploting resulting clusters
colors=['purple','red','yellow','black']
plt.figure(figsize=(10, 10))
plt.scatter(df[0], df[1], c=df['HR_labels'],cmap=matplotlib.colors.ListedColormap(colors),s=15)
plt.title('HR_labels',fontsize=20)
plt.ylabel = 'Actual Output'
plt.xlabel = 'predicted output'
plt.show()
```

## DB Scan Method

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(df[[0,1]])
distances, indices = nbrs.kneighbors(df[[0,1]])


from sklearn.cluster import DBSCAN
dbscan_opt = DBSCAN(eps=30,min_samples=6)
dbscan_opt.fit(df[[0,1]])

df['DBSCAN_opt_labels']= dbscan_opt.labels_
df['DBSCAN_opt_labels'].value_counts()

# ploting resulting clusters
colors=['purple','red','yellow','black']
plt.figure(figsize=(10, 10))
plt.scatter(df[0], df[1], c=df['DBSCAN_opt_labels'],cmap=matplotlib.colors.ListedColormap(colors),s=15)
plt.title('DBSCAN Clustering',fontsize=20)
plt.ylabel = 'Actual Output'
plt.xlabel = 'predicted output'
plt.show()
```
